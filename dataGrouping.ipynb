```
popDict={}
rockDict={}
hiphopDict={}
soulDict={}
edmDict={}
rbDict={}
alternativeDict={}
countryDict={}
jazzDict={}
metalDict={}
folkDict={}
punkDict={}

maleDict={}
femaleDict={}
OtherValue={}
for i in demoDict:
  array=demoDict[i]
  if array[1] == "Pop": #check if the demographics of the artist is "pop" or any genre
    popDict.update({i:i})
  elif array[1] == "Hip-Hop":
    hiphopDict.update({i:i})
  elif array[1] == "Rock":
    rockDict.update({i:i})
  elif array[1] == "Soul":
    soulDict.update({i:i})
  elif array[1] == "EDM":
    edmDict.update({i:i})
  elif array[1] == "R&B":
    rbDict.update({i:i})
  elif array[1] == "Alternative":
    alternativeDict.update({i:i})
  elif array[1] == "Country":
    countryDict.update({i:i})
  elif array[1] == "Jazz":
    jazzDict.update({i:i})
  elif array[1] == "Metal":
    metalDict.update({i:i})
  elif array[1] == "Folk":
    folkDict.update({i:i})
  elif array[1] == "Punk":
    punkDict.update({i:i})
 
  if array[0] == 'Male': #checks gender of the artist
    maleDict.update({i:i})
  elif array[0] == 'Female':
    femaleDict.update({i:i})
  elif array[0] == 'Other':
    OtherValue.update({i:i})





```
