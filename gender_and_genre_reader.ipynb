Iterates through names and finds the genre and gender of each artists and stores it into a dictionary and json.

```
import requests
from lxml.html import fromstring
import requests
from bs4 import BeautifulSoup

demographics ={}
for i in nameDict: #iterates through the keys and stores the name as "i"
  name=i
  url= f"https://en.wikipedia.org/wiki/{name.replace(' ', '_')}" #adds the name to the URL
  array=scrape(url) #calls function
  if array[1]==' ': #if it runs and there is no genre found, it checks a different wikipedia page
    url= f"https://en.wikipedia.org/wiki/{name.replace(' ', '_')}_(musician)"
    array=scrape(url)
    if array[1]==' ':
      url= f"https://en.wikipedia.org/wiki/{name.replace(' ', '_')}_(singer)"
      array=scrape(url)
      if array[1]==' ':
        url= f"https://en.wikipedia.org/wiki/{name.replace(' ', '_')}_(group)"
        array=scrape(url)
        if array[1]==' ':
          url= f"https://en.wikipedia.org/wiki/{name.replace(' ', '_')}_(band)"
          array=scrape(url)

  print(name, ': ', array) #prints each pair
  demographics.update({name: array}) #updates each pair to a dictionary with name as the key and the returning array as the value


with open("gender-and-genre.json","w") as cp: #writes to a json file
  json.dump(demographics,cp,indent=4)
```
